# 範例重點

如何使用亂數, 資料集來操作


```python
import matplotlib.pyplot as plt
import numpy as np

# 準備數據 ... 假設我要畫一個sin波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,180)
y = np.sin(x * np.pi / 180.0)

# 開始畫圖
plt.plot(x,y)
plt.xlim(0, 200)
plt.ylim(-1, 1.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('sin')
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 

#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)
plt.show()
```

```python
#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)
plt.show()
plt.savefig("Day018_HW.png",dpi=300,format="png")
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

#給定散點圖顏色
plt.scatter(X, Y, c = 'red', marker = 'o', alpha = 0.4, linewidth = 1.0, edgecolor = 'black')

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
