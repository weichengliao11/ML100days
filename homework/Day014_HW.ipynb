### 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。


```python
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text
data = json.loads(response)
```

```python
data = pd.DataFrame(data)
data.sort_values(by = ['subscriptionCount'], ascending = False)
data_csv = data.to_csv('data.csv', index = False)
data_csv
```
